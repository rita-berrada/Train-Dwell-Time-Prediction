```python
import pandas as pd
```

```python
# Load Data
events_df = pd.read_csv('Events Data.csv')
schedule_df = pd.read_csv('Schedule Data.csv')
```

```python
print(list(events_df.columns))
print(list(schedule_df.columns))
```

```python
events_df['TRN_DAY'].unique()
```

```python
# For events_df
events_df['TRN_DAY'] = events_df['TRN_DAY'].astype(str).str.zfill(2)

# For schedule_df
schedule_df['TRN_DAY'] = schedule_df['TRN_DAY'].astype(str).str.zfill(2)
```

```python
events_df['TRN_DAY'].unique()
```

```python
# Create Train ID for Events and Schedule Data
events_df['TRAIN_ID'] = (
    events_df['TRN_TYPE'].astype(str) +
    events_df['TRN_SYM'].astype(str) +
    events_df['TRN_SECT'].astype(str) +
    events_df['TRN_DAY'].astype(str) +
    events_df['TRN_PRTY'].astype(str)
)

schedule_df['TRAIN_ID'] = (
    schedule_df['TRN_TYPE'].astype(str) +
    schedule_df['TRN_SYM'].astype(str) +
    schedule_df['TRN_SECT'].astype(str) +
    schedule_df['TRN_DAY'].astype(str) +
    schedule_df['TRN_PRTY'].astype(str)
)
```

```python
events_df.head()
```

```python
events_df['TRN_SCH_DPT_DT'].unique()
```

```python
# Create length column
events_df['TRAIN_ID_LENGTH'] = events_df['TRAIN_ID'].str.len()
schedule_df['TRAIN_ID_LENGTH'] = schedule_df['TRAIN_ID'].str.len()

# Append date to length 11 TRAIN_IDs
events_df.loc[events_df['TRAIN_ID_LENGTH'] == 11, 'TRAIN_ID'] = (
    events_df['TRAIN_ID'] + events_df['TRN_SCH_DPT_DT'].astype(str)
)

schedule_df.loc[schedule_df['TRAIN_ID_LENGTH'] == 11, 'TRAIN_ID'] = (
    schedule_df['TRAIN_ID'] + schedule_df['TRN_SCH_DPT_DT'].astype(str)
)

# Update length column after appending
events_df['TRAIN_ID_LENGTH'] = events_df['TRAIN_ID'].str.len()
schedule_df['TRAIN_ID_LENGTH'] = schedule_df['TRAIN_ID'].str.len()


print(f"Events: {len(events_df)} rows")
print(f"Schedule: {len(schedule_df)} rows")
```

```python
events_df.head()
```

```python
schedule_df.head()
```

```python
# Keep only rows with length 21 (those that had dates appended)
events_df = events_df[events_df['TRAIN_ID_LENGTH'] == 21]
schedule_df = schedule_df[schedule_df['TRAIN_ID_LENGTH'] == 21]

# Drop the length column
events_df.drop('TRAIN_ID_LENGTH', axis=1, inplace=True)
schedule_df.drop('TRAIN_ID_LENGTH', axis=1, inplace=True)

print(f"Events remaining: {len(events_df)} rows")
print(f"Schedule remaining: {len(schedule_df)} rows")
```

```python
events_df['TRAIN_ID'].unique()
```

```python
# Remove all spaces
events_df['TRAIN_ID'] = events_df['TRAIN_ID'].str.replace(' ', '')
schedule_df['TRAIN_ID'] = schedule_df['TRAIN_ID'].str.replace(' ', '')
```

```python
events_df['TRAIN_ID'].unique()
```

```python
# Remove trailing spaces from STN_333 column in both dataframes

print(schedule_df['STN_333'].unique())
events_df['STN_333'] = events_df['STN_333'].str.strip()
schedule_df['STN_333'] = schedule_df['STN_333'].str.strip()
print(schedule_df['STN_333'].unique())
```

```python
# Merge events and schedule dataframes on common columns

columns_from_schedule = ['TRAIN_ID', 'STN_333', 'STN_ST', 'REQ_INSP', 'CREW_CHG', 'DPT_DIR']
merge_columns = ['TRAIN_ID', 'STN_333', 'STN_ST']

schedule_subset = schedule_df[columns_from_schedule]

merged_df = events_df.merge(schedule_subset,
                          on=merge_columns,
                          how='left')

print(f"Events shape: {events_df.shape}")
print(f"Schedule subset shape: {schedule_subset.shape}")
print(f"Merged shape: {merged_df.shape}")
print(f"\nMerged columns: {list(merged_df.columns)}")
merged_df.head()
```

```python
merged_df['TRAIN_ID'].nunique()
```

```python
# Create a new column that combines date and time
merged_df['EVT_DTTM'] = pd.to_datetime(merged_df['EVT_DT'] + ' ' + merged_df['EVT_TM'])
print(merged_df['EVT_DTTM'].dtype)
merged_df.head()
```

```python
# Get two df : one with respect to TA and one with respect to TD
columns_to_suffix = ['EVT_DT', 'EVT_TM', 'EVT_CD', 'PROC_DTTM', 'EVT_DTTM', 'DPT_DIR']

# Filter rows for TA
ta_df = merged_df[merged_df['EVT_CD'] == 'TA'].copy()

# Rename columns with TA suffix
ta_rename_dict = {col: f'{col}_TA' for col in columns_to_suffix}
ta_df.rename(columns=ta_rename_dict, inplace=True)

# Filter rows for TD
td_df = merged_df[merged_df['EVT_CD'] == 'TD'].copy()

# Rename columns with TD suffix
td_rename_dict = {col: f'{col}_TD' for col in columns_to_suffix}
td_df.rename(columns=td_rename_dict, inplace=True)
```

```python
ta_df.head()
```

```python
td_df.head()
```

```python
# Sort with the new column of event date and time
ta_df.sort_values(by= ['TRAIN_ID', 'EVT_DTTM_TA'], inplace=True)
td_df.sort_values(by= ['TRAIN_ID', 'EVT_DTTM_TD'], inplace=True)
```

```python
# Drop duplicates based on TRAIN_ID in each dataframe
print(f"Before dropping duplicates:")
print(f"TA dataframe: {ta_df.shape}")
print(f"TD dataframe: {td_df.shape}")

# Drop duplicates keeping the first occurrence
ta_df = ta_df.drop_duplicates(subset=['TRAIN_ID'], keep='first')
td_df = td_df.drop_duplicates(subset=['TRAIN_ID'], keep='last')

print(f"\nAfter dropping duplicates:")
print(f"TA dataframe: {ta_df.shape}")
print(f"TD dataframe: {td_df.shape}")
```

```python
ta_df.head()
```

```python
td_df.head()
```

```python
# Define merge columns
merge_columns = ['TRAIN_ID','TRN_TYPE','TRN_SECT', 'TRN_SYM','TRN_DAY','TRN_PRTY','TRN_SCH_DPT_DT',	 'STN_333','STN_ST','STN_SEQ_NBR', 'REQ_INSP','STN_TYPE_CD','CREW_CHG']

# Define which columns to take from td_df
columns_from_td = merge_columns + [col for col in td_df.columns if col.endswith('_TD')]

# Create subset of td_df with only needed columns
td_subset = td_df[columns_from_td]

# Left merge
final_df = ta_df.merge(td_subset,
                      on=merge_columns,
                      how='left')
```

```python
final_df=  final_df.dropna(subset=['PROC_DTTM_TA', 'PROC_DTTM_TD'])
```

```python
print(f"TA dataframe shape: {ta_df.shape}")
print(f"TD subset shape: {td_subset.shape}")
print(f"Final merged shape: {final_df.shape}")
```
